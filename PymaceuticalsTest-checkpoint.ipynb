```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress
from scipy.stats import sem
import numpy as np
from sklearn import datasets
from matplotlib.pyplot import figure



#Study data files
mouse_metadata_path = "C:/Users/DeferentAtom94/OneDrive/Desktop/Resources/Pandas_Challenge/Mouse_metadata.csv"
study_results_path = "C:/Users/DeferentAtom94/OneDrive/Desktop/Resources/Pandas_Challenge/Study_results.csv"

#Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#Combine the data into a single dataset
meta_study_complete = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])

#Display the data table for preview
meta_study_complete.head(10)
```

## Summary Statistics

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#================================================================================================================================

#Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#Mean of the tumor volume using the groupby function (drug regimen) as well as the mean function for average
#===========================================================================================================
mice_regi_mean = scrubbed_mouse_results.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
mice_regi_mean
```

```python
#Median of the tumor volume using the groupby function (drug regimen) as well as the median function
#===================================================================================================
mice_regi_median = scrubbed_mouse_results.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
mice_regi_median
```

```python
#Variance of the tumor volume using the groupby function (drug regimen) as well as the variance function
#=======================================================================================================
mice_regi_variance = scrubbed_mouse_results.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
mice_regi_variance
```

```python
#Standard Deviation of the tumor volume using the groupby function (drug regimen) as well as the Standard Deviation function
#===========================================================================================================================
mice_regi_std = scrubbed_mouse_results.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
mice_regi_std
```

```python
#Sem of the tumor volume of the tumor volume using the groupby function (drug regimen) as well as the SEM function
#===================================================================================================================
mice_regi_sem = scrubbed_mouse_results.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
mice_regi_sem
```

```python
#Generate a summary statistics table of mean, median, variance, standard deviation, and SEM
#of the tumor volume for each regimen
#=================================================================================================
summary_table = pd.DataFrame({"Mean": mice_regi_mean,
                              "Median": mice_regi_median,
                              "Variance": mice_regi_variance,
                              "Standard Deviation": mice_regi_std,
                              "SEM": mice_regi_sem})
#Print Summary Table
summary_table
```

```python
#Using the aggregation method, produce the same summary statistics in a single line
#=====================================================================================================
single_line_summary = scrubbed_mouse_results.groupby('Drug Regimen')

#Creating another summary table using the aggregation function
#This function returns a single value from multiple values taken as input which are grouped together on certain criteria.

summary_table_v2 = single_line_summary.agg(['mean','median','var','std','sem'])["Tumor Volume (mm3)"]

summary_table_v2
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr = st.pearsonr(avg_capomulin_volume['Weight (g)'],avg_capomulin_volume['Tumor Volume (mm3)'])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")
```

```python
#Calculate the correlation coefficient and linear regression model for mouse weight
# and average tumor volume for the Capomulin regimen
#=====================================================================================
x_values = avg_capomulin_volume['Weight (g)']
y_values = avg_capomulin_volume['Tumor Volume (mm3)']

(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept

print(f"slope:{slope}")
print(f"intercept:{intercept}")
print(f"rvalue (Correlation coefficient):{rvalue}")
print(f"pandas (Correlation coefficient):{corr}")
print(f"stderr:{stderr}")

line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

print(line_eq)
```

```python
#Add the linear regression equation and line to plot
#===========================================================================================
fig1, ax1 = plt.subplots(figsize=(15, 10))
plt.scatter(x_values,y_values,s=175, color="royalblue")
plt.plot(x_values,regress_values,"r-")
plt.title('Regression Plot for Average Tumor Volume Versus Mouse Weight  ',fontsize =22)
plt.xlabel('Weight(g)',fontsize =16)
plt.ylabel('Average Tumore Volume (mm3)',fontsize =16)
ax1.annotate(line_eq, xy=(20, 40), xycoords='data',xytext=(0.8, 0.95), textcoords='axes fraction',horizontalalignment='right', verticalalignment='top',fontsize=30,color="darkblue")

print(f"The r-squared is: {rvalue**2}")

#Show your work
plt.show()
```
